```julia
using PyPlot
using Random
```

```julia
rng = MersenneTwister()

function rand_init(size)
    orbitals = bitrand(rng, size)
    return orbitals
end 
```

```julia
function run_sim(sys_size, n_steps, beta, mu)
    orbitals_init = rand_init(sys_size)
    orbitals_mc = zeros(sys_size)
    
    for step_cnt = 0: n_steps
        r_orb_index = rand(1:sys_size)
        
        if orbitals_init[r_orb_index] == 1
            sign = -1.0
        else
            sign = 1.0 
        end
        
        delta_e = r_orb_index 
        accept = min(1.0, exp( -beta *(delta_e - mu)*sign ))
        reject = rand(Float64)
        
        if accept >= reject
            if orbitals_init[r_orb_index] == 0
                orbitals_init[r_orb_index] = 1
            else
                orbitals_init[r_orb_index] = 0
            end
        end 
        
        orbitals_mc += orbitals_init
        
    end
    
    return orbitals_mc/n_steps
end
result_sim = run_sim(100, 100000, 1/10, 50)
```

```julia
# run DMRG calculation and return observer (energy truncation to speed up simulations)
function run_theory(sys_size, beta, my)

    orbitals_theory = []
    for orb = 0: sys_size
        push!(orbitals_theory, 1.0/(1.0 + exp(beta*(orb-my))) ) 
    end
    
    return orbitals_theory
end
result_theory = run_theory(100, 1/10, 50.0)
```

```julia
#pygui(true)

fig = figure(figsize = (9, 6))

xlabel("Energy")
ylabel("Av Occupation number")

plot( result_sim, marker = "o", markersize = 5.0, color = "blue", label="MC Sim")
plot( result_theory, marker = "o", markersize = 5.0, color = "red", label="Theory")

legend()

tight_layout()
title("Monte Carlo sampled Fermi–Dirac distribution")
show()  
```
